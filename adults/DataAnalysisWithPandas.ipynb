```python
import pandas as pd
data=pd.read_csv("adult.csv")
data
```

```python
#8. Drop all The Missing Values
data.isnull().sum()*100/len(data)

'''The output shows 0.0 for all columns, indicating no missing values. 
This is because I had already handled and dropped the missing values in the dataset before running this check.'''
```

```python
data.dropna(how='any',inplace=True)
data.shape

```

```python
#9. Check For Duplicate Data and Drop Them
# Check for duplicate rows
duplicate_count = data.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
'''The dataset might contain duplicate rows, so we check for duplicates first. 
In this case, if duplicates exist, they are counted and shown. 
After identifying the duplicates, we remove them from the dataset.'''
# Drop duplicate rows
data.drop_duplicates(inplace=True)

# Check the shape of the dataset after removing duplicates
print(f"Shape after removing duplicates: {data.shape}")
'''At this point, the dataset should no longer contain duplicate rows. 
However, if the output shows no changes, it's because the dataset might not have had any duplicates 
or they were already removed in earlier steps.'''
```
